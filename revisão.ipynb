```python
import random


class TwistWord:
    def __init__(self):
        self.tentativas = 5
        self.venceu = False
        self.dificuldades = {
            "facil": 5,
            "medio": 3,
            "dificil": 1,
        }
        self.temas = [
            {
                "nome": "Cidades",
                "palavras": [
                    "São Paulo",
                    "Rio de Janeiro",
                    "Belo Horizonte",
                    "Porto Alegre",
                    "Curitiba",
                    "Salvador",
                    "Fortaleza",
                    "Brasília",
                    "Goiânia",
                    "Joinville",
                ],
            },
            {
                "nome": "Cores",
                "palavras": [
                    "Azul",
                    "Vermelho",
                    "Amarelo",
                    "Verde",
                    "Roxo",
                    "Preto",
                    "Branco",
                ],
            },
            {
                "nome": "Animais",
                "palavras": [
                    "Cachorro",
                    "Gato",
                    "Leão",
                    "Tigre",
                    "Girafa",
                    "Cavalo",
                    "Galinha",
                    "Elefante",
                    "Macaco",
                    "Pinguim",
                    "Urso",
                    "Cobra",
                    "Coelho",
                    "Rato",
                ],
            },
            {
                "nome": "Times",
                "palavras": [
                    "São Paulo",
                    "Corinthians",
                    "Palmeiras",
                    "Santos",
                    "Flamengo",
                    "Vasco",
                    "Fluminense",
                    "Botafogo",
                    "Grêmio",
                    "Internacional",
                ],
            },
        ]

    def pick_word(self):
        print("Escolha um tema:")
        for i, tema in enumerate(self.temas):
            print(f"{i + 1}. {tema['nome']}")

        tema_escolhido = self.get_valid_input(
            "Digite o número do tema desejado: ", 1, len(self.temas)
        )
        tema_selecionado = self.temas[tema_escolhido - 1]
        palavra_secreta = random.choice(tema_selecionado["palavras"])

        return palavra_secreta

    def twist_word(self, word):
        word_list = list(word)
        random.shuffle(word_list)
        return "".join(word_list)

    def mostrar_motivacao(self):
        frases = [
            "Você consegue, tente novamente",
            "Não desista, tente novamente",
            "Não foi dessa vez, tente novamente",
            "Não foi dessa vez, tente novamente",
        ]
        print(random.choice(frases))

    def mostrar_parabens(self):
        frases = [
            "Parabens, você venceu!",
            "Incrível, você ganhou!",
        ]
        print(random.choice(frases))

    def get_valid_input(self, prompt, min_val, max_val):
        while True:
            try:
                value = int(input(prompt))
                if min_val <= value <= max_val:
                    return value
                else:
                    print("Opção inválida. Digite novamente.")
            except ValueError:
                print("Entrada inválida. Digite novamente.")

    def jogar(self):
        while True:
            print("Escolha a dificuldade:")
            for dificuldade, tentativas in self.dificuldades.items():
                print(f"{dificuldade.capitalize()} - {tentativas} tentativas")

            dificuldade_escolhida = input(
                "Digite o nome da dificuldade desejada: "
            ).lower()
            self.tentativas = self.dificuldades.get(dificuldade_escolhida, 5)

            while not self.tentativas == 0 and not self.venceu:
                palavra_secreta = self.pick_word()
                palavra_torcida = self.twist_word(palavra_secreta)

                tentativa = str(input(f"[{palavra_torcida}] QUAL SEU CHUTE? --> "))

                if tentativa == palavra_secreta:
                    self.venceu = True
                else:
                    self.tentativas -= 1
                    self.mostrar_motivacao()

            if self.tentativas == 0:
                print(f"Você perdeu, a palavra era {palavra_secreta}")
            elif self.venceu:
                self.mostrar_parabens()

            self.venceu = False
            play_again = input("Deseja jogar novamente? (sim/não): ")
            if play_again.lower() != "sim":
                break
            else:
                self.tentativas = 0


jogo = TwistWord()

jogo.jogar()
```
